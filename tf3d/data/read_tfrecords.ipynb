This is an example of reading .tfrecords files. We start from importing the modules required:

```python
import tensorflow as tf

import numpy as np
import IPython.display as display
```

We open a .tfrecords file of Waymo dataset and load it as a TFRecord Dataset:

```python
filenames = "waymo_dataset/original_tfrecords_test-00000-of-00528.tfrecords"
raw_dataset = tf.data.TFRecordDataset(filenames)
raw_dataset
```

Since the data file is too large, we print the output into a file. 

```python
outputFile = open('original_tfrecords_test-00000-of-00528.tfrecords.txt', 'w')

for raw_record in raw_dataset.take(1):
  example = tf.train.Example()
  example.ParseFromString(raw_record.numpy())
  print(example, file = outputFile)

outputFile.close()
```

```python
outputFile = open('original_tfrecords_test-00000-of-00528.tfrecords.txt', 'r')

for i in range(1,50):
   print(outputFile.readline())

outputFile.close()
```

Read the first 50 line of a .tfrecords file. We can see the key and values of each feature in tfrecords.
